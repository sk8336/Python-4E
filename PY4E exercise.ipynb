```python
#Count words using dictionary
word = 'brontosaurus'
d = dict()
for c in word:
    if c not in d:
        d[c]=1
    else:
        d[c] = d[c]+1
print(d)
```

```python
# We can use get to write our histogram loop more concisely. Because the get
# method automatically handles the case where a key is not in a dictionary, we can
# reduce four lines down to one and eliminate the if statement.

word = 'brontosaurus'
d = dict()
print(d)
for c in word:
    d[c] = d.get(c,0) + 1
    print(d[c],c)
print(d)
```

```python
import string
fname = input('Enter the file name: ')
try:
    fhand = open(fname)
except:
    print('File cannot be opened:', fname)
    exit()
counts = dict()
for line in fhand:
    line = line.rstrip()
    line = line.translate(line.maketrans('', '', string.punctuation))
    line = line.lower()
    words = line.split()
    for word in words:
        if word not in counts:
            counts[word] = 1
        else:
            counts[word] += 1
print(counts)
```
